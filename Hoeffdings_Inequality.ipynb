# Hoeffding's Inequality

Suppose we have a container of marbles. Some are red, some are green, and we don't know the ratio. We draw $N$ independent samples with replacement from the container. 

We define the ratio of $\nu = \frac{\text{# red marbles drawn}}{N}$ 

$\mu$ is the actual ratio of red marbles to all marbles in the container.

We want to know if $\nu$ is close to $\mu$. Formally,

$$\lvert \nu - \mu \lvert < \epsilon$$.

Hoeffding's Inequality gives us a lower bound:

$$\Large P( \lvert \nu - \mu \lvert > \epsilon ) \le 2e^{-2N\epsilon^2}$$

Very powerful as the bound does not depend on $\mu$!

However, as $\epsilon$ gets smaller you need exponentially larger $N$ to get the same error bounds because of the $\epsilon^2$ term.

```python
import math

def hoeffding(epsilon, N):
    return 2*math.exp(-2 * N * (epsilon**2))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
eps = 0.1
N = 1000

eps_range = np.arange(1.e-3, .1, .001)
hs = []
for eps in eps_range:
    hs.append(hoeffding(eps, N))
plt.plot(eps_range, hs);    
```

```python
ns = np.arange(1, 1000)
hs = []
for N in ns:
    hs.append(hoeffding(0.05, N))
plt.plot(ns, hs);
```
