# Bayesian Estimation

Bayes. Variable or parameter X, and data Y.

$$P(X|Y) = \frac{P(Y|X)\times P(X)}{P(Y)} = \frac{P(Y|X)\times P(X)}{\int P(Y|X)\times P(X) \, \mathrm{d}X}$$

$P(Y|X)$ is the *likelihood*.

$P(X)$ is the *prior*.

$P(Y)$ is the *evidence*.


# MAP

Estimation requires finding most probable value of P(X|Y). Logical to find the most probably value - its maximum. *Maximu a posteriori* (MAP) estimate is the value of x that maximixes the posterior density:

$$ \max\limits_x P(X|Y)$$

Find this by differentiating, set to 0, and solve:

$$\nabla_x P(X|Y)\bigr\rvert_{X=\hat{X}_{MAP}}=0$$

where $\rvert_x$ means *evaluated* at x.

The *gradient vector* $\nabla_x \in R^{N_X \times 1}$

$$\nabla_X := \begin{bmatrix}\frac{\partial}{\partial X_1}\cdots\frac{\partial}{\partial X_{N_X}}\end{bmatrix}^\mathsf{T}$$

(not sure what $N_X$ is)


$\ln$ is useful since we often work with exponential densities. Logarithm is monotonically increasing so maximums hap[en at the same points. 


$$\nabla_x \ln P(X|Y)\bigr\rvert_{X=\hat{X}_{MAP}}=0$$

Apply Bayes rule:
$$\ln P(X|Y) = \ln P(Y|X) + \ln P(X) - \ln P(Y)$$

P(Y) is not a function of X, so map equation is

$$\nabla_x \ln P(X|Y)\bigr\rvert_{X=\hat{X}_{MAP}} = \nabla_x  (\ln P(Y|X) + \ln P(X))\bigr\rvert_{X=\hat{X}_{MAP}} = 0$$
